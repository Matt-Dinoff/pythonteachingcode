### Guessing a letter A-Z  
**check_guess()** takes 2 string arguments: **letter and guess** (both expect single alphabetical character)   
    - if guess is not an alpha character print invalid and return False
    - test and print if guess is "high" or "low" and return False
    - test and print if guess is "correct" and return True

```python
# [ ] create check_guess()
def check_guess(letter,guess):
    guess = str(guess)
    if guess.isalpha():
        #guess is alphabetical letters
        if len(guess)==1:
            #guess is one character
            #TODO check the guess
            if guess == letter:
                print("Correct")
                return True
            elif guess > letter:
                print("Low")
                return False
            elif guess < letter:
                print("High")
                return False
            else:
                print("unhandled exception")
                return False
        else:
            print("Invalid - more than one character")
            return False
    else:
        print("Invalid - not alphabetical")
        return False
    
def letter_game(correct_answer):
    this_guess = input("Guess the letter (provide one letter): ")
    if True == check_guess(correct_answer,this_guess):
        print("On the first try! What a mensch.")
    else:
        #second guess
        this_guess = input("Guess the letter (provide one letter): ")
        if True == check_guess(correct_answer,this_guess):
            print("On the second try! What a mensch.")
        else:
        #third guess
            this_guess = input("Guess the letter (provide one letter): ")
            if True == check_guess(correct_answer,this_guess):
                print("On the third try! Sortta slow.")
            else:
                print("You lose.")

letter_game("A")
```

```python
# [ ] call check_guess with user input
check_guess("A",input("Guess the letter (provide one letter): "))

```

### Letter Guess
**create letter_guess() function that gives user 3 guesses**
- takes a letter character argument for the answer letter
- gets user input for letter guess  
- calls check_guess() with answer and guess
- End letter_guess if 
    - check_guess() equals True, return True  
    - or after 3 failed attempts, return False

```python
# [ ] create letter_guess() function, call the function to test
def letter_guess(correct_answer):
    guess = input("Guess the letter (provide one letter): ")
    if True == check_guess(correct_answer,guess):
        print("Correct! First try!")
        return True
    else:
        guess = input("Guess the letter (provide one letter): ")
        if True == check_guess(correct_answer,guess):
            print("Nice job! Second try!")
            return True
        else:
            if True == check_guess(correct_answer,guess):
                print("Phew! Third try!")
                return True
            else:
                print("You lose.")
                return False
        

letter_guess("A")

```

### Rainbow colors
ask for input of a favorite rainbow color first letter: ROYGBIV  

Using `if`, `elif`, and `else`:  
- print the color matching the letter  
    - R = Red  
    - O = Orange  
    - Y = Yellow  
    - G = Green
    - B = Blue
    - I = Indigo
    - V = Violet
    - else print "no match"


```python
# [ ] complete rainbow colors
color_choice = input("What is your favorite rainbow color first letter?")
if color_choice == "R":
    print("R = Red")
elif color_choice == "O":
    print("O = Orange")
elif color_choice == "Y":
    print("Y = Yellow")
elif color_choice == "G":
    print("G = Green")
elif color_choice == "B":
    print("B = Blue")
elif color_choice == "I":
    print("I = Indigo")
elif color_choice == "V":
    print("V = Violet")
else:
    print("No match.")
```

```python
# [ ] make the code above into a function rainbow_color() that has a string parameter, 
# get input and call the function and return the matching color as a string or "no match" message.
# Call the function and print the return string.
def rainbow_color(color_choice):
    if color_choice == "R":
        return "R = Red"
    elif color_choice == "O":
        return "O = Orange"
    elif color_choice == "Y":
        return "Y = Yellow"
    elif color_choice == "G":
        return "G = Green"
    elif color_choice == "B":
        return "B = Blue"
    elif color_choice == "I":
        return "I = Indigo"
    elif color_choice == "V":
        return "V = Violet"
    else:
        return "No match."
     
color_choice = input("What is your favorite rainbow color first letter?")

rainbow_color(color_choice)
```

**Create function age_20() that adds or subtracts 20 from your age for a return value based on current age** (use `if`) 
- call the funtion with user input and then use the return value in a sentence  
example `age_20(25)` returns **5**: 
> "5 years old, 20 years difference from now"

```python
# [ ] complete age_20()
age = int(input("Please enter your current age: "))
def age_20(current_age):
    if current_age > 20:
        return str(current_age - 20) + " years old, 20 years difference from then."
    if current_age < 20:
        return str(current_age + 20) + " years old, 20 years difference from now."
    
age_20(age)
```

**create a function rainbow_or_age that takes a string argument**
- if argument is a digit return the value of calling age_20() with the str value cast as **`int`** 
- if argument is an alphabetical character return the value of calling rainbow_color() with the str
- if neither return FALSE

```python
# [ ]  create rainbow_or_age()
user_input_age_color = input("Enter your age or the first letter of your favorite color of the rainbow: ")
def rainbow_or_age(user_input_age_color):
    if user_input_age_color.isdigit():
        return age_20(int(user_input_age_color))
    elif user_input_age_color.isalpha():
        return rainbow_color(user_input_age_color)
    else:
        return "False"
    
rainbow_or_age(user_input_age_color)

    
```
